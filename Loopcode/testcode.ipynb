```python
import pandas as pd
import numpy as np
import json
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from geopy.distance import geodesic
import folium
from sklearn.metrics.pairwise import cosine_similarity
import matplotlib.pyplot as plt

# Pad naar de bestanden
user_profiles_path = '../Data/Loopdata/kleinebestanden/run_gebruikersprofielen.json'
routes_path = '../Data/Loopdata/kleinebestanden/routes_volledig.json'

user_profiles_df = pd.read_json(user_profiles_path, lines=True)
routes_df = pd.read_json(routes_path, lines=True)

# Preprocessing van de routegegevens
routes_df['mean_altitude'] = routes_df['altitude'].apply(np.mean)
routes_df['mean_latitude'] = routes_df['latitude'].apply(np.mean)
routes_df['mean_longitude'] = routes_df['longitude'].apply(np.mean)
user_profiles_df[['most_common_lat', 'most_common_lon']] = pd.DataFrame(user_profiles_df['average_location'].tolist(), index=user_profiles_df.index)

# Normaliseer de kenmerken met StandardScaler
scaler = StandardScaler()
user_features = scaler.fit_transform(user_profiles_df[['average_speeds', 'average_altitude_changes', 'average_distances', 'average_condition_value']])
route_features = scaler.fit_transform(routes_df[['altitude_change', 'distance', 'mean_altitude', 'average_condition_value']])

# Pas PCA toe op de gebruikers- en routekenmerken
pca_user = PCA(n_components=0.95)
pca_route = PCA(n_components=0.95)
user_pca = pca_user.fit_transform(user_features)
route_pca = pca_route.fit_transform(route_features)

# Voeg de PCA-resultaten toe aan de DataFrames
user_profiles_df[['PCA1', 'PCA2']] = user_pca[:, :2]
routes_df[['PCA1', 'PCA2']] = route_pca[:, :2]

# Bepaal gewichten voor gebruikerskenmerken
gewichten = {}
for index, row in user_profiles_df.iterrows():
    user_id = row['userId']
    gewichten[user_id] = {
        'PCA1': row['weight_std_altitude_change'], 
        'PCA2': row['weight_std_distance'],
        'distance_to_user': 1.0}

# Definieer aanbevelingsfunctie voor routes
def recommend_routes_for_user_corrected(user_id, user_profiles_df, routes_df, top_n=10):
    user_data = user_profiles_df[user_profiles_df['userId'] == user_id]
    user_location = (user_data['most_common_lat'].iloc[0], user_data['most_common_lon'].iloc[0])

    # Bereken de afstand tussen de gebruiker en elke route
    routes_df['distance_to_user'] = routes_df.apply(lambda row: geodesic(user_location, (row['mean_latitude'], row['mean_longitude'])).kilometers, axis=1)

    # Cosine similarity berekenen
    # Binnen de recommend_routes_for_user_corrected functie

    # Cosine similarity berekenen
    def calculate_cosine_similarity(row):
        user_values = user_data[['PCA1', 'PCA2']].values[0]
        route_values = row[['PCA1', 'PCA2', 'distance_to_user']]
        # Voeg een element toe voor de afstand tot de gebruiker en pas het gewicht toe
        gewogen_user_values = np.append(user_values, 1.0) * [gewichten[user_id]['PCA1'], gewichten[user_id]['PCA2'], gewichten[user_id]['distance_to_user']]
        return cosine_similarity([gewogen_user_values], [route_values])[0][0]

    routes_df['similarity_score'] = routes_df.apply(calculate_cosine_similarity, axis=1)


    # routes_df['similarity_score'] = routes_df.apply(calculate_cosine_similarity, axis=1)

    # Selecteer de routes met de hoogste cosine similarity scores
    recommended_routes = routes_df.sort_values(by='similarity_score', ascending=False).head(top_n)

    # Visualiseer de resultaten
    plt.figure(figsize=(10, 6))
    plt.hist(routes_df['similarity_score'], bins=30, color='blue', alpha=0.7)
    plt.title(f'Histogram van Similarity Scores voor Gebruiker {user_id}')
    plt.xlabel('Similarity Score')
    plt.ylabel('Aantal Routes')
    plt.grid(True)
    plt.show()

    # Creëer een kaart met de locatie van de gebruiker en de routes
    map_obj = folium.Map(location=user_location, zoom_start=12)
    folium.Marker(user_location, popup=f"UserID: {user_id}").add_to(map_obj)

    colors = ["red", "blue", "green", "purple", "orange"]
    for index, route in recommended_routes.iterrows():
        route_coords = list(zip(route['latitude'], route['longitude']))
        route_color = colors[index % len(colors)]
        folium.PolyLine(route_coords, color=route_color, weight=2.5, opacity=1).add_to(map_obj)

    display(map_obj)  # Toon de kaart direct
    return recommended_routes

# Test de functie voor een specifieke gebruiker
sample_user_id = 3545637
recommended_routes = recommend_routes_for_user_corrected(sample_user_id, user_profiles_df, routes_df)
recommended_routes.head(10)
```

```python
import pandas as pd
import numpy as np
from sklearn.neighbors import NearestNeighbors
from geopy.distance import geodesic
import folium
import itertools

# Pad naar de bestanden
user_profiles_path = '../Data/Loopdata/kleinebestanden/run_gebruikersprofielen.json'
routes_path = '../Data/Loopdata/kleinebestanden/routes_volledig.json'

# Laad de gebruikersprofielen en routes
user_profiles_df = pd.read_json(user_profiles_path, lines=True)
routes_df = pd.read_json(routes_path, lines=True)

# Preprocessing van de routegegevens
routes_df['mean_latitude'] = routes_df['latitude'].apply(np.mean)
routes_df['mean_longitude'] = routes_df['longitude'].apply(np.mean)
user_profiles_df[['most_common_lat', 'most_common_lon']] = pd.DataFrame(user_profiles_df['average_location'].tolist(), index=user_profiles_df.index)

# KNN clustering voor gebruikersprofielen
def cluster_user_profiles(user_profiles_df, n_neighbors=5):
    clustering_features = ['average_speeds', 'average_altitude_changes', 'average_distances', 'average_condition_value', 'most_common_lat', 'most_common_lon']
    knn = NearestNeighbors(n_neighbors=n_neighbors)
    knn.fit(user_profiles_df[clustering_features])
    return knn

# Aanbevelingen op basis van soortgelijke gebruikers
def recommend_based_on_similar_users(user_id, user_profiles_df, routes_df, knn, top_n=10):
    clustering_features = ['average_speeds', 'average_altitude_changes', 'average_distances', 'average_condition_value', 'most_common_lat', 'most_common_lon']
    distances, indices = knn.kneighbors([user_profiles_df.loc[user_profiles_df['userId'] == user_id, clustering_features].values[0]])
    similar_users = user_profiles_df.iloc[indices[0]].userId.tolist()

    # Haal routes op die populair zijn bij soortgelijke gebruikers
    popular_routes = routes_df[routes_df['userId'].isin(similar_users)]

    # Tel de frequentie van elke route bij soortgelijke gebruikers
    route_popularity = popular_routes['id'].value_counts().rename('route_popularity')

    # Voeg populariteitsinformatie toe aan routes_df
    routes_with_popularity = routes_df.merge(route_popularity, left_on='id', right_index=True, how='left')
    routes_with_popularity['route_popularity'] = routes_with_popularity['route_popularity'].fillna(0)

    # Filter en sorteer routes op basis van populariteit
    recommended_routes = routes_with_popularity[routes_with_popularity['id'].isin(popular_routes['id'])]
    recommended_routes = recommended_routes.sort_values(by='route_popularity', ascending=False).head(top_n)
    
    return recommended_routes
# Functie om routes en gebruikersinfo op een kaart te visualiseren
def visualize_recommended_routes(user_id, user_profiles_df, recommended_routes):
    # Haal de locatie van de gebruiker op
    user_data = user_profiles_df[user_profiles_df['userId'] == user_id].iloc[0]
    user_location = (user_data['most_common_lat'], user_data['most_common_lon'])

    # Maak een Folium kaart
    map_obj = folium.Map(location=user_location, zoom_start=12)

    # Voeg een marker toe voor de gebruiker
    user_popup_info = f"UserID: {user_id}<br>" \
                      f"Gemiddelde Snelheid: {user_data['average_speeds']}<br>" \
                      f"Gemiddelde Hoogteverandering: {user_data['average_altitude_changes']}<br>" \
                      f"Gemiddelde Afstand: {user_data['average_distances']}<br>" \
                      f"Gemiddelde Conditiescore: {user_data['average_condition_value']}"
    folium.Marker(
        user_location,
        popup=user_popup_info,
        icon=folium.Icon(color="red", icon="info-sign"),
    ).add_to(map_obj)

    # Definieer een lijst met kleuren
    colors = itertools.cycle(["red", "blue", "green", "purple", "orange", "darkred", "lightred", "beige", "darkblue", "darkgreen", "cadetblue", "darkpurple", "white", "pink", "lightblue", "lightgreen", "gray", "black", "lightgray"])

    # Voeg de aanbevolen routes toe aan de kaart
    for _, route_data in recommended_routes.iterrows():
        route_coords = list(zip(route_data['latitude'], route_data['longitude']))
        route_color = next(colors)
        route_popup_info = f"Route ID: {route_data['id']}<br>" \
                           f"Gebruiker ID: {route_data['userId']}<br>" \
                           f"Route Afstand: {route_data['distance']}<br>" \
                           f"Route Hoogteverandering: {route_data['altitude_change']}<br>"
        folium.PolyLine(route_coords, color=route_color, weight=5, opacity=0.7, popup=route_popup_info).add_to(map_obj)

    return map_obj
# Gebruik de functies
knn_users = cluster_user_profiles(user_profiles_df)
# sample_user_id = 3545637  
sample_user_id = 5255745
recommended_routes = recommend_based_on_similar_users(sample_user_id, user_profiles_df, routes_df, knn_users)

# Genereer en toon de kaart
map_obj = visualize_recommended_routes(sample_user_id, user_profiles_df, recommended_routes)
display(map_obj)

# Toon de aanbevolen routes met populariteitsgegevens
print(recommended_routes[['id', 'route_popularity']].head())
```

```python
import pandas as pd
import numpy as np
import json
from sklearn.preprocessing import MinMaxScaler, StandardScaler
from sklearn.neighbors import NearestNeighbors
from geopy.distance import geodesic
import folium
from scipy.spatial.distance import euclidean
from sklearn.metrics.pairwise import cosine_similarity
import matplotlib.pyplot as plt
import seaborn as sns
import itertools

# Pad naar de bestanden
user_profiles_path = '../Data/Loopdata/kleinebestanden/run_gebruikersprofielen.json'
routes_path = '../Data/Loopdata/kleinebestanden/routes_volledig.json'

# Laad de gebruikersprofielen en routes
user_profiles_df = pd.read_json(user_profiles_path, lines=True)
routes_df = pd.read_json(routes_path, lines=True)

# Sla de oorspronkelijke gegevens op in nieuwe kolommen voordat je ze normaliseert
user_profiles_df['orig_average_speeds'] = user_profiles_df['average_speeds']
user_profiles_df['orig_average_altitude_changes'] = user_profiles_df['average_altitude_changes']
user_profiles_df['orig_average_distances'] = user_profiles_df['average_distances']
user_profiles_df['orig_average_condition_value'] = user_profiles_df['average_condition_value']

routes_df['orig_altitude_change'] = routes_df['altitude_change']
routes_df['orig_distance'] = routes_df['distance']
routes_df['orig_mean_altitude'] = routes_df['altitude'].apply(np.mean)
routes_df['orig_average_condition_value'] = routes_df['average_condition_value']

# Preprocessing van de routegegevens
routes_df['mean_latitude'] = routes_df['latitude'].apply(np.mean)
routes_df['mean_longitude'] = routes_df['longitude'].apply(np.mean)
routes_df['mean_altitude'] = routes_df['altitude'].apply(np.mean)
user_profiles_df[['most_common_lat', 'most_common_lon']] = pd.DataFrame(user_profiles_df['average_location'].tolist(), index=user_profiles_df.index)

# Normaliseer de kenmerken
scaler = StandardScaler()
user_profiles_df[['average_speeds', 'average_altitude_changes', 'average_distances', 'average_condition_value']] = scaler.fit_transform(user_profiles_df[['average_speeds', 'average_altitude_changes', 'average_distances', 'average_condition_value']])
routes_df[['altitude_change', 'distance', 'mean_altitude', 'average_condition_value']] = scaler.fit_transform(routes_df[['altitude_change', 'distance', 'mean_altitude', 'average_condition_value']])

# Bepaal gewichten voor gebruikerskenmerken
gewichten = {}
for index, row in user_profiles_df.iterrows():
    user_id = row['userId']
    altitude_weight = row['weight_std_altitude_change']
    distance_weight = row['weight_std_distance']
    condition_weight = row['weight_std_condition_value']
    gewichten[user_id] = {
        'average_altitude_change': altitude_weight, 
        'average_distance': distance_weight, 
        'average_condition_value': condition_weight,
        'distance_to_user': 1.0}

# Aanbevelingsfunctie voor routes
def recommend_routes_for_user_corrected(user_id, user_profiles_df, routes_df, top_n=10):
    user_data = user_profiles_df[user_profiles_df['userId'] == user_id]
    user_location = (user_data['most_common_lat'].iloc[0], user_data['most_common_lon'].iloc[0])


    # Bereken de afstand tussen de gebruiker en elke route
    def calculate_distance_to_user(row):
        route_location = (row['mean_latitude'], row['mean_longitude'])
        distance_to_user = geodesic(user_location, route_location).kilometers
        return distance_to_user

    routes_df['distance_to_user'] = routes_df.apply(calculate_distance_to_user, axis=1)

    # Functie om de cosine similarity te berekenen
    def calculate_cosine_similarity(row):
        user_values = user_data[['average_altitude_changes', 'average_distances', 'average_condition_value']].values[0]
        route_values = row[['altitude_change', 'distance', 'distance_to_user', 'average_condition_value']]

        # Bereken de gewogen waarden
        gewogen_user_values = np.array([
            user_values[0] * gewichten[user_id]['average_altitude_change'],
            user_values[1] * gewichten[user_id]['average_distance'],
            user_values[2] * gewichten[user_id]['average_condition_value'],
            gewichten[user_id]['distance_to_user']])

        gewogen_route_values = np.array([
            route_values['altitude_change'],
            route_values['distance'],
            route_values['average_condition_value'],
            route_values['distance_to_user']])

        # Cosine similarity berekenen
        return cosine_similarity([gewogen_user_values], [gewogen_route_values])[0][0]


    # Pas de cosine similarity functie toe en sla de score op
    routes_df['similarity_score'] = routes_df.apply(calculate_cosine_similarity, axis=1)
# Selecteer de routes met de hoogste cosine similarity scores
    recommended_routes = routes_df.sort_values(by='similarity_score', ascending=False).head(top_n)

    return recommended_routes

# KNN clustering voor gebruikersprofielen
def cluster_user_profiles(user_profiles_df, n_neighbors=5):
    clustering_features = ['average_speeds', 'average_altitude_changes', 'average_distances', 'average_condition_value']
    knn = NearestNeighbors(n_neighbors=n_neighbors)
    knn.fit(user_profiles_df[clustering_features])
    return knn

# Aanbevelingen op basis van soortgelijke gebruikers
def recommend_based_on_similar_users(user_id, user_profiles_df, routes_df, knn, top_n=10):
    clustering_features = ['average_speeds', 'average_altitude_changes', 'average_distances', 'average_condition_value']
    distances, indices = knn.kneighbors([user_profiles_df.loc[user_profiles_df['userId'] == user_id, clustering_features].values[0]])
    similar_users = user_profiles_df.iloc[indices[0]].userId.tolist()
    
    # Haal routes op die populair zijn bij soortgelijke gebruikers
    popular_routes = routes_df[routes_df['userId'].isin(similar_users)]
    recommended_routes = popular_routes.sort_values(by='similarity_score', ascending=False).head(top_n)
    return recommended_routes

# Hybride aanbevelingsfunctie
# Hybride aanbevelingsfunctie
def recommend_routes_hybrid(user_id, user_profiles_df, routes_df, knn, top_n=30):
    recommended_routes_content_based = recommend_routes_for_user_corrected(user_id, user_profiles_df, routes_df, top_n)
    recommended_routes_similar_users = recommend_based_on_similar_users(user_id, user_profiles_df, routes_df, knn, top_n)
    
    # Combineer de aanbevelingen en voorkom duplicaten
    combined_recommendations = pd.concat([recommended_routes_content_based, recommended_routes_similar_users]).drop_duplicates(subset=['id'])

    return combined_recommendations.head(top_n)

# Aangepaste Hybride aanbevelingsfunctie met kaartgeneratie
# Aangepaste Hybride aanbevelingsfunctie met kaartgeneratie
def recommend_routes_hybrid_with_map(user_id, user_profiles_df, routes_df, knn, top_n=30):
    recommended_routes_hybrid = recommend_routes_hybrid(user_id, user_profiles_df, routes_df, knn, top_n)

    # Haal de locatie van de gebruiker op
    user_data = user_profiles_df[user_profiles_df['userId'] == user_id].iloc[0]
    user_location = (user_data['most_common_lat'], user_data['most_common_lon'])

    # Maak een popup-informatie voor de gebruiker met oorspronkelijke waarden
    user_popup_info = f"UserID: {user_id}<br>" \
                      f"Gemiddelde Snelheid: {user_data['orig_average_speeds']}<br>" \
                      f"Gemiddelde Hoogteverandering: {user_data['orig_average_altitude_changes']}<br>" \
                      f"Gemiddelde Afstand: {user_data['orig_average_distances']}<br>" \
                      f"Gemiddelde Conditiescore: {user_data['orig_average_condition_value']}"

    # Maak een Folium kaart
    map_obj = folium.Map(location=user_location, zoom_start=12)

    # Voeg een marker toe voor de gebruiker
    folium.Marker(
        user_location,
        popup=user_popup_info,
        icon=folium.Icon(color="red", icon="info-sign"),
    ).add_to(map_obj)

    # Definieer een lijst met kleuren en gebruik itertools.cycle om oneindig door de kleuren te itereren
    colors = itertools.cycle(["red", "blue", "green", "purple", "orange", "darkred", "lightred", "beige", "darkblue", "darkgreen", "cadetblue", "darkpurple", "white", "pink", "lightblue", "lightgreen", "gray", "black", "lightgray"])

    # Voeg de aanbevolen routes toe aan de kaart met verschillende kleuren
    for route in recommended_routes_hybrid.iterrows():
        _, route_data = route
        route_coords = list(zip(route_data['latitude'], route_data['longitude']))
        route_color = next(colors)  # Wijs een kleur toe aan elke route
        route_popup_info = f"Route ID: {route_data['id']}<br>" \
                           f"Gebruiker ID: {route_data['userId']}<br>" \
                           f"Route Afstand: {route_data['orig_distance']}<br>" \
                           f"Route Hoogteverandering: {route_data['orig_altitude_change']}<br>" \
                           f"Conditiescore: {route_data['orig_average_condition_value']}<br>" \
                           f"Similariteitsscore: {route_data['similarity_score']}"
        folium.PolyLine(route_coords, color=route_color, weight=5, opacity=0.7, popup=route_popup_info).add_to(map_obj)

    return recommended_routes_hybrid, map_obj
# Gebruik de functies
# Gebruik de functies
knn_users = cluster_user_profiles(user_profiles_df)
sample_user_id = 3545637
recommended_routes_hybrid, map_obj = recommend_routes_hybrid_with_map(sample_user_id, user_profiles_df, routes_df, knn_users)
display(map_obj)

plt.figure(figsize=(10, 6))
plt.hist(recommended_routes_hybrid['similarity_score'], bins=30, color='blue', alpha=0.7)
plt.title(f'Histogram van Similarity Scores voor Gebruiker {sample_user_id}')
plt.xlabel('Similarity Score')
plt.ylabel('Aantal Routes')
plt.grid(True)
plt.show()

recommended_routes_hybrid.head()
```
