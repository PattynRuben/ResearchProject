```python
import json
import requests
from collections import Counter
from concurrent.futures import ThreadPoolExecutor
import threading

# Functie om JSON-data uit een bestand te laden
def load_json_data(file_path):
    print(f"Laden van data uit {file_path}")
    with open(file_path, 'r') as file:
        return [json.loads(line) for line in file]

# Functie om bestaande IDs uit het uitvoerbestand te laden
def load_existing_ids(file_path):
    try:
        print(f"Laden van bestaande IDs uit {file_path}")
        with open(file_path, 'r') as file:
            return set(json.loads(line.split(': ')[0]) for line in file)
    except FileNotFoundError:
        print(f"Bestand niet gevonden: {file_path}. Geen bestaande IDs geladen.")
        return set()

# Functie om de Overpass API te bevragen
def query_overpass_api(coordinates, session):
    print(f"Bevragen van de Overpass API voor coördinaten: {coordinates[0]} ...")
    coordinate_str = ' '.join([f"{lat} {lon}" for lat, lon in coordinates])
    query = f"""
    [out:json];
    (
      way(poly:"{coordinate_str}")["highway"];
      way(poly:"{coordinate_str}")["amenity"];
      way(poly:"{coordinate_str}")["cycleway"];
      way(poly:"{coordinate_str}")["footway"];
      way(poly:"{coordinate_str}")["natural"];
      way(poly:"{coordinate_str}")["landuse"];
      node(poly:"{coordinate_str}")["tourism"="viewpoint"];
    );
    out body;
    """
    url = "http://overpass-api.de/api/interpreter"
    response = session.post(url, data={'data': query})
    if response.status_code == 200:
        print("API-verzoek succesvol.")
        return response.json()
    else:
        print(f"Fout bij API-verzoek: Statuscode {response.status_code}")
        return None

def process_route(json_obj, session, existing_ids, output_file):
    if json_obj['id'] not in existing_ids:
        print(f"Verwerken van route met ID: {json_obj['id']}")
        coordinates = list(zip(json_obj['latitude'], json_obj['longitude']))
        data = query_overpass_api(coordinates, session)
        route_features = Counter()

        if data:
            for element in data['elements']:
                if 'tags' in element:
                    tags = element['tags']
                    route_features.update(tags.items())

        with output_file_lock:
            output_file.write(f"{json_obj['id']}: {json.dumps(route_features)}\n")
            output_file.flush()  # Flush na het schrijven naar het bestand
            print(f"Route met ID {json_obj['id']} verwerkt en opgeslagen.")


# Hoofdscript
file_path = '../Data/Loopdata/kleinebestanden/routes_volledig.json'
output_file_path = '../Data/Loopdata/RouteSurfaceData4.txt'

existing_ids = load_existing_ids(output_file_path)
json_data = load_json_data(file_path)

with requests.Session() as session, open(output_file_path, 'a') as output_file:
    output_file_lock = threading.Lock()
    print("Begin van de verwerking van routes...")
    
    with ThreadPoolExecutor(max_workers=10) as executor:
        futures = [executor.submit(process_route, json_obj, session, existing_ids, output_file) for json_obj in json_data]

    # Wachten op voltooiing van alle taken
    for future in futures:
        future.result()

print("Data opgeslagen in", output_file_path)
```
